```python
import DEC

import matplotlib.pyplot as plt
import numpy as np

base_vertex = [0.0, 0.0, 0.0]
r = 3 / np.pi

l1 = 1.5
l2 = 1

length = 3
n_bottoms = int(length // 0.1)
print(n_bottoms)

angle_guess = np.pi / 4

lower_angle_guess = 0.001
upper_angle_guess = 100

while upper_angle_guess - lower_angle_guess > 1e-2:
    bottom_vertices = np.array(
        [
            [
                ((1 - theta) * l1 + theta * l2) * np.cos(theta * angle_guess),
                0,
                ((1 - theta) * l1 + theta * l2) * np.sin(theta * angle_guess),
            ]
            for theta in np.linspace(0, 1, n_bottoms)
        ]
    )

    # compute the total length between bottom vertices
    computed_length = 0
    for i in range(1, n_bottoms):
        computed_length += np.linalg.norm(bottom_vertices[i] - bottom_vertices[i - 1])

    if computed_length < length:
        lower_angle_guess = angle_guess
        angle_guess = (angle_guess + upper_angle_guess) / 2
    else:
        upper_angle_guess = angle_guess
        angle_guess = (angle_guess + lower_angle_guess) / 2

vertices = np.array([base_vertex] + bottom_vertices.tolist())
print(np.zeros((n_bottoms + 1)).shape)
print(np.linspace(0, 0.2, n_bottoms + 1).shape)
print(vertices.shape)
vertices += np.stack(
    (
        np.zeros((n_bottoms + 1)),
        np.linspace(0, 0.2, n_bottoms + 1),
        np.zeros((n_bottoms + 1)),
    )
).T
faces = [[0, i, i + 1] for i in range(1, n_bottoms)]

mesh = DEC.Mesh(vertices, faces)
mesh.dump_to_JSON(
    f"AA_squiggle_flattened_{length}.json",
    {"X": np.array([0] + np.linspace(0, 0, n_bottoms).tolist())},
    folder="../../produced_solutions",
)
print(n_bottoms)
```

```python
import DEC

import matplotlib.pyplot as plt
import numpy as np

base_vertex = [0.0, 0.0, 0.0]
r = 3 / np.pi


n_bottoms = int(length // 0.01)


angle = np.pi / 2
left = l1 * np.array([np.cos(0), 0.0, np.sin(0)])
right = l2 * np.array([np.cos(angle), 0.0, np.sin(angle)])

dist = np.linalg.norm(left - right)

freq = 1
height = min(l1, l2) / 10

from scipy.integrate import quad

# binary search to find the frequency of a sine wave over

minfreq = 0.01
maxfreq = 1000
k = 0
while maxfreq - minfreq > 1e-2:
    k += 1
    if k > 100:
        break
    freq = (minfreq + maxfreq) / 2
    lengtth = quad(
        lambda x: np.sqrt((freq * np.cos(freq * x) * height) ** 2 + 1), 0, dist
    )
    if lengtth[0] < length:
        minfreq = freq
    else:
        maxfreq = freq

bottom_vertices = np.array(
    [
        ((1 - t) * left + (t * right))
        + np.array(
            [
                0,
                np.sin(freq * t * dist) * height,
                0,
            ]
        )
        for t in np.linspace(0, 1, n_bottoms)
    ]
)


# split bottom vertices in 13 contiguous parts
splits = 4
idxs = np.linspace(0, n_bottoms, splits + 1).astype(int)
# idxs = np.linspace(0, n_bottoms, n_bottoms).astype(int)
print(idxs)

dir = left - right

faces = []
vertices = []
colors = []
for i in range(splits + 1):
    print(i)
    if i == 0:
        continue

    start = len(vertices)
    in_split = bottom_vertices[idxs[i - 1] : idxs[i] + 1]

    # print(idxs[i - 1] : idxs[i] + 1)

    mean = np.mean(in_split, axis=0)
    in_split = (in_split - mean) * 0.9 + mean
    new_vertices = [
        base_vertex - dir * (i / splits) * min(l1, l2) * 0.05
    ] + in_split.tolist()
    # new_vertices += np.array([0, (i % 2) * min(l1, l2) * 0.1, 0])
    vertices += new_vertices
    faces += [[start, i, i + 1] for i in range(start + 1, start + len(in_split))]
    colors += [1.0 * i % 2] * (len(in_split) + 1)

print(vertices)

print(len(colors))
print(len(vertices))
print(colors)


mesh = DEC.Mesh(np.array(vertices), faces)
mesh.dump_to_JSON(
    f"AA_squiggle_split_folded_{length}, {n_bottoms}.json",
    {"X": colors},
    folder="../../produced_solutions",
)
```
